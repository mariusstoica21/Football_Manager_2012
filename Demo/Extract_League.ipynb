<h1>Extract_League</h1>

<b>Description</b>:

The objective of the current script is to read the data from the "Leagues/England/Premier_League_2012.txt" and "Leagues/England/Championship_2012.txt" files, to parse the content, and save into "Dataframes/Premier_League_2012.csv" and "Dataframe/Championship_2012.csv".

The process was repeated for all the leagues in the game. Therefore, the .txt files containing the footballers name, attributes, age, position and value were saved into .csv files that contained the parsed information from dataframes.

The dataframes from the .csv file would eventually be loaded into a single dataframe with all the footballers from Football Manager 2012.

<h2>England First League 2012</h2>

```python
import pandas as pd
import numpy as np
```

Open the "Leagues/England/Premier_League_2012.txt", read the content and parse by the "|" delimiter. It is not the cleanest parsing process, it could have been written in cleaner Python code, but it did it's job for all the leagues extracted.

```python
# read by default 1st sheet of an excel file
df_England_1_2012 = pd.read_fwf("Leagues/England/Premier_League_2012.txt",encoding='unicode_escape')

newDf_England_1_2012 = df_England_1_2012.iloc[1::2]

r,c = newDf_England_1_2012.shape

print(r,c)

dfPlayers_England_1_2012 = pd.DataFrame(
    {'Name': [], 
     'Acc': [],
     'Agi': [],
     'Bal' : [],
     'Jum' : [],
     'Nat' : [],
     'Pac' : [],
     'Sta' : [],
     'Str' : [],
     'Agg' : [],
     'Ant' : [],
     'Bra' : [],
     'Cmp' : [],
     'Cnt' : [],
     'Cre' : [],
     'Dec' : [],
     'Det' : [],
     'Fla' : [],
     'Inf' : [],
     'Off' : [],
     'Pos' : [],
     'Tea' : [],
     'Wor' : [],
     'Cor' : [],
     'Cro' : [],
     'Dri' : [],
     'Fin' : [],
     'Fir' : [],
     'Fre' : [],
     'Hea' : [],
     'Lon' : [],
     'LTh' : [],
     'Mar' : [],
     'Pas' : [],
     'Pen' : [],
     'Tac' : [],
     'Tec' : [],
     'Aer' : [],
     'Cmd' : [],
     'Com' : [],
     'Ecc' : [],
     'Han' : [],
     'Kic' : [],
     'One' : [],
     'Ref' : [],
     'Rus' : [],
     'Pun' : [],
     'Thr' : [],
     'Position': [],
     'Age' : [],
     'Value' : []
    })

#For each row of the file
for i in range(r):
    line = newDf_England_1_2012.iloc[i]
    print(line)
    row = []
    noAttributes = dfPlayers_England_1_2012.shape[1]
    for l in range(noAttributes):
        row.append(0)
    for j in range(c): 
            if (j == 1): 
                lst = line.values[j].split("|")
                row[0] = lst[2]
                row[1] = lst[3]
                row[2] = lst[4]
                row[3] = lst[5]
                row[4] = lst[6]
                row[5] = lst[7]
                row[6] = lst[8]
                row[7] = lst[9]
                row[8] = lst[10]   
            if (j == 3):
                lst = line.values[j].split("|")
                row[9] = lst[3]
                row[10] = lst[4]
                row[11] = lst[5]
                row[12] = lst[6]
                row[13] = lst[7]
                row[14] = lst[8]
                row[15] = lst[9]
                row[16] = lst[10]
                row[17] = lst[11]   
                row[18] = lst[12]
                row[19] = lst[13]
                row[20] = lst[14]
                row[21] = lst[15]
                row[22] = lst[16]
            if (j == 5):
                lst = line.values[j].split("|")
                row[23] = lst[3]
                row[24] = lst[4]
                row[25] = lst[5]
                row[26] = lst[6]
                row[27] = lst[7]
                row[28] = lst[8]
                row[29] = lst[9]
                row[30] = lst[10]
                row[31] = lst[11]   
                row[32] = lst[12]
                row[33] = lst[13]
                row[34] = lst[14]
                row[35] = lst[15]
                row[36] = lst[16]  
            if (j == 7):
                lst = line.values[j].split("|")
                row[37] = lst[3]
                row[38] = lst[4]
                row[39] = lst[5]
                row[40] = lst[6]
                row[41] = lst[7]
                row[42] = lst[8]
                row[43] = lst[9]
                row[44] = lst[10]
                row[45] = lst[11]   
                row[46] = lst[12]
                row[47] = lst[13]
            if (j == 8):
                lst = line.values[j].split("|")
                row[48] = lst[4]
                row[49] = lst[9]
                row[50] = lst[10]
    dfPlayers_England_1_2012.loc[len(dfPlayers_England_1_2012)] = row
```

735 players are extracted from the English Premier League

```python
dfPlayers_England_1_2012.shape
```

Out of the 735 players extracted, there were some cells that did not contain information. Therefore, those cells were replaced with NaN and removed from the dataframe. 
There are other techniques to insert values in those cells, like mean, median, average, but most of the empty cells belonged to players that had "not so accurate stats", therefore the final decision was to remove them.

```python
dfPlayers_England_1_2012 = dfPlayers_England_1_2012.replace(r'^\s*$', np.nan, regex=True)

dfPlayers_England_1_2012.dropna(axis=0, inplace=True)
```

```python
dfPlayers_England_1_2012.shape
```

Print the dataframe of the English Premier League 2012 season.

```python
dfPlayers_England_1_2012
```

Save the dataframe to "Dataframes/Premier_League_2012.csv". All of the dataframes resulted from parsing every league of the game will eventually be concatenated into a single dataframe.

```python
# Export to CSV
file_path = 'Dataframes/Premier_League_2012.csv'
dfPlayers_England_1_2012.to_csv(file_path, index=False)
```
